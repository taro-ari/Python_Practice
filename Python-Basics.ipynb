```python
x =['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'] 
print(x[0],' ',x[-8])
print(x[::-1])
print(x[6:2:-1])
print(x[::2])
print(x[2:6], ' ', x[2:-2])
print(x[1:],' ',x[-7:])
print(x[:])
```

```python
import numpy as np
x = np.array([[[1],[2],[3]], [[4],[5],[6]]]) 
print (x.shape) 
```

```python
attention = 1
y = x[:,attention,:]
print(y, ' ', y.shape)
```

```python
z = np.squeeze(x, axis=2)
print(z, ' ', z.shape)
```

```python
w = x.flatten() # row-major (order='C') if order='F' (column-major)
print(w,' ', w.shape)
```

```python
u = np.reshape(x,(3,2,1))
print(u)
```

```python
x = np.array([-2, 15, 4, 0, -7, 3])
print(x)
```

```python
x = np.matrix(np.array([6.5, -2.1, 0.0, 1.9])).T
print(x)
```

```python
x = np.matrix ( #行列は行ごと
    [[1,0,0,-1],
     [0,2,-1,1],
     [-1,-1,0,1],
     [0,-2,2,-1] ]) 
print(x)
```

```python
x = ["東京", "ニューヨーク", "パリ", "ロンドン", "北京"]
print(x)
```

```python
import pandas as pd
x = pd.DataFrame({'A': np.array([2,5,4,1,9]),
    'B': pd.Categorical(["東京", "ニューヨーク", "パリ", "ロンドン", "北京"])})
print(x)
```

```python
x = np.array([6.5, -2.1, 0.0, 1.9])
y = np.array([1.1, -2.8, -1.0, 0.0])
z  = np.dot(x,y)
print (z)
print (np.sqrt(np.dot(x,x))) # xの大きさ
print (np.sqrt(np.dot(y,y))) # yの大きさ
```
