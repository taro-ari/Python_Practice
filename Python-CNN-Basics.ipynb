# 畳込みの概要

<pre>
畳込みでは、適当なサイズのカーネル窓（例 3x3（2次元の場合）)と、一般に多数のフィルタを
準備しておき、入力されたデータ（例　2次元画像）に、カーネル窓を端から端まで、stride幅で
（2次元の場合、左右上下方向に）スキャンしながらフィルタ関数を適用しては、その結果を次の層に出力
していきます。プーリングする場合は、あるフィルタを提要して得られたカーネル窓ごとに、
プーリング演算（例　平均や最大値）を実行し、該当する次の層の位置に結果を書き出します。
すなわち、多層パーセプトロン（MLP）と違い、全結合しません。このため、MLPと異なり、ある程度
位置情報を層にわたって伝搬することが可能です。
</pre>

# 畳込みの種類

<pre>
1次元畳込み(Conv1D)：主としてテキストや時系列データの処理に使います
2次元畳込み(Conv2D)：主として画像処理に使います
3次元畳込み(Conv3D)：主として動画処理や3Dデータ処理等に使います
2次元畳込み＋再帰型(ConvLSTM): 主として動画処理や3Dデータ処理に使います
</pre>

<b>注意</b> CPUだけの状況では、2次元以上の畳込みは非常に遅い（体感で10～50倍
ほど遅い）ので、これらの畳込みは、GPUメモリのあるマシンで実行することを
強く推奨します。
