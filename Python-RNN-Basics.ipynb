# 時刻tでの隠れ層の状態
$h_{t} = \phi(h_{t-1},x_{t})$

再帰型ニューラルネットワークは、大きく、３つの重み行列（パラメータ）を有する。
入力状態の$U$, 隠れ層の状態の$W$, 出力の$V$である。

\begin{array}{l}
i\left( t \right) = {\rm{sigmoid}}\left( {{W_i}h\left( {t - 1} \right) + {U_i}x\left( t \right)} \right)\\
f\left( t \right) = {\rm{sigmoid}}\left( {{W_f}h\left( {t - 1} \right) + {U_f}x\left( t \right)} \right)\\
o\left( t \right) = {\rm{sigmoid}}\left( {{W_o}h\left( {t - 1} \right) + {U_o}x\left( t \right)} \right)\\
g\left( t \right) = \tanh \left( {{W_g}h\left( {t - 1} \right) + {U_g}x\left( t \right)} \right)\\
c\left( t \right) = \left( {c\left( {t - 1} \right) \otimes f\left( t \right)} \right) \oplus \left( {g\left( t \right) \otimes i\left( t \right)} \right)\\
h\left( t \right) = \tanh \left( {c\left( t \right)} \right) \otimes o\left( t \right)
\end{array}
